## Delete useless columns

```python
data.drop(columns="RowNumber",axis=1,inplace=True)
data.drop(columns="Surname",axis=1,inplace=True)
data.drop(columns="CustomerId",axis=1,inplace=True)
```

```python
data.head()
```

## Create a variable 'Purchased'

Determine whether the customer made the purchase based on the three factors : Num Of Products / Point Earned / Satisfaction Score

```python
data['Purchased'] = data.apply(
    lambda row: 1 if (row['NumOfProducts'] > 1 and row['Point Earned'] > 0 and row['Satisfaction Score'] > 3) else 0, axis=1
)
```

```python
data.head(5)
```

```python
data.describe()
```

## Convert all categorical columns

```python
from sklearn.preprocessing import LabelEncoder , OneHotEncoder
```

```python
label_encoder = LabelEncoder()

data['Gender'] = label_encoder.fit_transform(data['Gender'])
```

```python
geography = data[['Geography']]
card_type = data[['Card Type']]

encoder = OneHotEncoder(sparse_output=False, drop='first')

geography_encoded = encoder.fit_transform(geography)
geography_encoded_df = pd.DataFrame(geography_encoded, columns=encoder.get_feature_names_out())

card_type_encoded = encoder.fit_transform(card_type)
card_type_encoded_df = pd.DataFrame(card_type_encoded, columns=encoder.get_feature_names_out())

data_prep = pd.concat([data.drop(columns=['Geography']), geography_encoded_df], axis=1)

data_prep = pd.concat([data_prep.drop(columns=['Card Type']), card_type_encoded_df], axis=1)
```
